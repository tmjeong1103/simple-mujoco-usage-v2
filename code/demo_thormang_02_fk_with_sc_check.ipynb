## Forward Kinematics of `Thormang`

```python
import numpy as np
np.set_printoptions(precision=2)
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
from mujoco_parser import MuJoCoParserClass
print ("Done.")
```

## Parse motion file (.csv)

```python
motion_folder_path = '../asset/thormang_rilab/motion_csv_file/original/'
motion_name = 'cmu_40_11_wait_for_bus.csv'
```

```python
motion_path = motion_folder_path + motion_name
print(motion_path)
```

```python
q = np.loadtxt(motion_path)
motion_tick = np.size(q[:,1])
n_rev_joint = np.size(q[1,:])
print('total motion tick: %d'%motion_tick)
print('total number of revolute joint: %d'%n_rev_joint)
```

```python
q_mod = np.zeros((motion_tick,n_rev_joint))
q_mod[:,:19] = q[:,:19]
q_mod[:,19:21] = q[:,30:32]
q_mod[:,21:33] = q[:,19:31]
```

## Forward kinematics with collision check for `Thormang`

```python
env = MuJoCoParserClass(name='thormang_rilab',rel_xml_path='../asset/thormang_rilab/thormang_rilab.xml',VERBOSE=True)
```

```python
env.init_viewer(TERMINATE_GLFW=True,INITIALIZE_GLFW=True,window_width=0.5,window_height=0.5)
env.set_max_sec(max_sec=10.0)
env.reset()
i_idx = 0
while env.IS_ALIVE():
    q = q_mod[i_idx,:]
    env.forward(q_pos=q_mod[i_idx,:])
    for c_idx in range(env.sim.data.ncon):
        contact    = env.sim.data.contact[c_idx]
        p_contact  = contact.pos
        bodyname1  = env.body_idx2name(env.sim.model.geom_bodyid[contact.geom1])
        bodyname2  = env.body_idx2name(env.sim.model.geom_bodyid[contact.geom2])
        label      = '%s-%s'%(bodyname1,bodyname2)
        env.add_marker(pos=p_contact,radius=0.1,color=np.array([1,0,0,0.5]),label=label)
        print("=============================================")
        print("*Warning* Collision tick: %d" %(i_idx+1))
        print("*Warning* Collision pair of joint: %s & %s" %(bodyname1,bodyname2))
    if i_idx != motion_tick-1:
        i_idx += 1
        env.render(cam_lookat=[-0.1,-0.3,1],RENDER_ALWAYS=True)
    else:       
        env.terminate_viewer()
print("Done.")
```
